# Punto 1:
Adicional al promedio calculado en el punto anterior, calcule el mínimo, el máximo, los cuartiles, la desviación estándar y el coeficiente de variación y realice el conteo de hogares. 

```python
def q1(x):
    return x.quantile(0.25)

def q2(x):
    return x.quantile(0.75)

def cv(x):
    return np.std(x) / np.mean(x) * 100


resumen = {'c10_valor': ['count','min','mean', q1, 'median', q2,'std', 'max', cv]}
```

```python
consulta1=punto0.query("c1 == 3").groupby(['nom_loc', 'b12'], as_index=False).agg(resumen)
```

# 5.	
¿Cuál es el hogar con mayor hacinamiento en Bogotá? (Calcule el cociente entre el número de personas sobre el número de cuartos). Cuál es el directorio de ese hogar. (DIRECTORIO_HOG).

```python
punto4=punto4.assign(hacinamiento = lambda x: x.c18 / x.c20_cuartos).sort_values('hacinamiento', ascending=False)
punto4.head()
```

# 6. En qué tipo de construcción se presentan las mayores tasas promedio de hacinamiento?

```python
punto6_b = b.filter(['directorio','b12'])
punto6_c = c.filter(['directorio','c18','c20_cuartos',])
punto6=punto6_b.merge(punto4_c, on='directorio', how='inner')
punto6=punto6.merge(a, on='directorio', how='inner')
punto6=punto6.merge(localidad, on='localidad', how='inner')
punto6=punto6.assign(hacinamiento = lambda x: x.c18 / x.c20_cuartos).sort_values('hacinamiento', ascending=False)
punto6.head()
```

```python
punto6.groupby(['b12'], as_index=False).agg({'hacinamiento':'mean'}).sort_values(['hacinamiento'],ascending=False)
```

Las casas con Otra forma de tenencia (posesión sin título, ocupante de hecho,
Pase a 9 propiedad colectiva, etc.) son las que más hacinamiento presentan.

# 7 	¿Cuál tipo de vivienda (B12) que presenta los mayores promedios en las tasas de hacinamiento en las localidades de Ciudad Bolívar y Usme?

```python
punto7=punto6.query("nom_loc in ['Ciudad Bolivar','Usme']").groupby(['b12'], as_index=False).agg({'hacinamiento':'mean'}).sort_values(['hacinamiento'],ascending=False)
punto7
```
